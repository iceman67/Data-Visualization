```python
import numpy as np
```

```python
data는 2차원 배열 2개의 열로 구성, 각 열은  x 축과 y 축의 값으로 사용됨 
```

```python
data = np.array([[1,2,3],[4,500,9000]])
```

```python
import matplotlib.pyplot as plt
```

```python
# x 축 [1,2,3]
# y 축 [4,500,9000]
plt.scatter(data[0,:], data[1,:])


plt.plot(data[0,:], data[1,:])
plt.xlabel("x axis")
plt.ylabel("y axis")
#plt.yscale("log")
plt.show()
```

```python
plt.bar(data[0,:], data[1,:])
plt.plot(data[0,:], data[1,:])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
#plt.yscale("log") 
```

```python
plt.barh(data[0,:], data[1,:])
plt.xlabel("x axis")
plt.ylabel("y axis")
#plt.yscale("log")
plt.show()
```

```python
plt.bar(data[0,:], data[1,:])
plt.plot(data[0,:], data[1,:], 'x--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks(data[0,:])
plt.yticks(data[1,:])
plt.show()
#plt.yscale("log") 
```

```python
plt.bar(data[0,:], data[1,:])
plt.plot(data[0,:], data[1,:], 'x--')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks(data[0,:])
plt.yticks(data[1,:])

plt.yscale("log") 
plt.show()
```
